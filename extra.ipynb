```python
data = pd.read_csv('./datasets/GasPricesinBrazil_2004-2019_preprocessado.csv')
```

```python
# pode também ser assim com LOC(lambda) encadeado (Extra 1)
posto_sp_rj_gas_maior_2_1 = (
    data.loc[(data.ESTADO == 'SAO PAULO') | (data.ESTADO == 'RIO DE JANEIRO')]
    .loc[lambda df: df['PRODUTO'] == 'GASOLINA COMUM']
    .loc[lambda df: df['PREÇO MÉDIO REVENDA'] > 2.0]
)
posto_sp_rj_gas_maior_2_1
```

```python
# pode também ser assim com LOC(lambda) e ISIN encadeado (Extra 2)
# o uso do lambda faz o encadeamento, assim pegando o que ja tem inicial do data
posto_sp_rj_gas_maior_2_2 = (
    data.loc[data['ESTADO'].isin(['SAO PAULO', 'RIO DE JANEIRO'])]
    .loc[lambda df: df['PRODUTO'].isin(['GASOLINA COMUM'])]
    .loc[
        lambda df: df['PREÇO MÉDIO REVENDA'].isin(
            df['PREÇO MÉDIO REVENDA'][df['PREÇO MÉDIO REVENDA'] > 2.0]
        )
    ]
)
posto_sp_rj_gas_maior_2_2
```

```python
# pode ser assim com PIPE(lambda) (ainda não estudado) (Extra 3)
posto_sp_rj_gas_maior_2_3 = (
    data.pipe(
        lambda df: df[
            (df['ESTADO'] == 'SAO PAULO') | (df['ESTADO'] == 'RIO DE JANEIRO')
        ]
    )
    .pipe(lambda df: df[df['PRODUTO'] == 'GASOLINA COMUM'])
    .pipe(lambda df: df[df['PREÇO MÉDIO REVENDA'] > 2.0])
)

posto_sp_rj_gas_maior_2_3
```

```python
# pode ser assim com PIPE(lambda) e ISIN (ainda não estudado) (Extra 4)
# o uso do lambda faz o encadeamento, assim pegando o que ja tem inicial do data
posto_sp_rj_gas_maior_2_4 = (
    data.pipe(lambda df: df[df['ESTADO'].isin(['SAO PAULO', 'RIO DE JANEIRO'])])
    .pipe(lambda df: df[df['PRODUTO'].isin(['GASOLINA COMUM'])])
    .pipe(
        lambda df: df[
            df['PREÇO MÉDIO REVENDA'].isin(
                df['PREÇO MÉDIO REVENDA'][df['PREÇO MÉDIO REVENDA'] > 2.0]
            )
        ]
    )
)
posto_sp_rj_gas_maior_2_4
```
